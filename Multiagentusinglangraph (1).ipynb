```python
def add(a: float, b: float) -> float:
    """Add two numbers."""
    return a + b

def multiply(a: float, b: float) -> float:
    """Multiply two numbers."""
    return a * b
```

```python
from langgraph_supervisor import create_supervisor
from langchain.agents import create_agent

math_agent = create_agent(
    model=model,
    tools=[add, multiply],
    name="math_expert"
)
```

```python
from langchain.agents import create_agent

research_agent = create_agent(
    model=model,
    tools=[search_duckduckgo],
    name="research_expert"
)
```

```python
from langgraph_supervisor import create_supervisor

supervisor_agent = create_supervisor(
    model=model,
    agents=[math_agent, research_agent],
    system_prompt="You are a supervisor capable of coordinating between a math_expert and a research_expert. Based on the user's request, decide which agent should act next to fulfill the request. If a specific task is asked that requires both agents, you can call them sequentially. If a research question is asked, use the research_expert. If a math question is asked, use the math_expert."
)
```

```python
workflow = create_supervisor(
    [research_agent, math_agent],
    model=model,
    prompt=(
        "You are a team supervisor managing a research expert and a math expert. "
        "For current events, use research_agent. "
        "For math problems, use math_agent."
    )
)
```

```python
# Compile and run
app = workflow.compile()

result = app.invoke({
    "messages": [
        {
            "role": "user",
            "content": "what is quantum computing?"
        }
    ]
})
```
